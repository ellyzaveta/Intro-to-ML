```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def input_csv(file_name):
    csv_reader = csv.reader(open(file_name))
    next(csv_reader)
    data = []
    for line in csv_reader:
        line.pop(0)
        data.append(line)
    return data
```

```python
def visualization(data, title, color):
    plt.title(title)
    plt.plot([float(line[0]) for line in data], [float(line[1]) for line in data], color)
```

```python
def general_visualization(train_data, test_data, train_color, test_color, b0, b1, title):
    plt.title(title)
    plt.plot([float(line[0]) for line in train_data], [float(line[1]) for line in train_data], train_color,
             [float(line[0]) for line in test_data], [float(line[1]) for line in test_data], test_color,
             [0, 1], [linear_regression(b0, b1, 0), linear_regression(b0, b1, 1)])
```

```python
def linear_regression(b0, b1, x):
    return b0 + b1 * x
```

```python
def MSE_loss_function(b0, b1, data):
    total_error = 0
    for line in data:
        total_error += (float(line[1]) - (linear_regression(b0, b1, float(line[0])))) ** 2
    return total_error / float(len(data))
```

```python
def gradient_descent(b0, b1, data, l_r):
    b0_gradient = 0
    b1_gradient = 0
    n = float(len(data))
    for line in data:
        b1_gradient += -(2/n) * float(line[0]) * (float(line[1]) - (linear_regression(b0, b1, float(line[0]))))
        b0_gradient += -(2/n) * (float(line[1]) - (linear_regression(b0, b1, float(line[0]))))
    b0 = b0 - l_r * b0_gradient
    b1 = b1 - l_r * b1_gradient
    return [b0, b1]
```

```python
def train(tolerance, l_r, data):
    b0 = float(abs(np.random.randn(1)))
    b1 = float(abs(np.random.randn(1)))
    loss = MSE_loss_function(b0, b1, data)
    delta = loss
    epoch_num = 0
    while tolerance < delta:
        b0, b1 = gradient_descent(b0, b1, data, l_r)
        print("epoch %d: b0 = %f; b1 = %f; loss function value = %f;" % (epoch_num, b0, b1, loss))
        delta = loss - MSE_loss_function(b0, b1, data)
        loss = MSE_loss_function(b0, b1, data)
        epoch_num += 1
    return [b0, b1]
```

```python
train_data = input_csv('lab_1_train.csv')
train_data
```

```python
visualization(train_data, "train data", "yo")
```

```python
tolerance = 0.000001
l_r = 0.01
b0, b1 = train(tolerance, l_r, train_data)
```
